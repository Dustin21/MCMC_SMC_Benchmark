```julia
means = [-2.0, 2.0, -5.0, -15.0]
sds = [1.0, 0.5, 0.5, 0.5]
weights = [1.0, 0.5, 0.5, 0.5]
function mix_gaussians(x, means = means, sds = sds, weights = weights, alpha = 1.0)
    l = length(means);
    out = 0;
    for i = 1:l
        out = out + weights[i] * exp(-(x + means[i]).^2 / (2*sds[i]))
    end
    return out.^(alpha).*exp(-(x).^2/2).^(1-alpha)
end

function twod(x1, x2 = 1.0, alpha = 1.0)
    m1 = [-2.0, 2.0, -5.0, -8.0]
    sd1 = [1.0, 0.5, 0.5, 0.5]
    w1 = [1.0, 0.5, 0.3, 0.5]
    m2 = [2.0, -8.0, -5.0, 7.0]
    sd2 = [0.5, 1.5, 1.2, 0.3]
    w2 = [2.0, 0.5, 1.0, 0.5]
    return mix_gaussians(x1, m1, sd1, w1, alpha) * mix_gaussians(x2*(x1/2), m2, sd2, w2, alpha) 
end
```
