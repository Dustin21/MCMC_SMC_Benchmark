```python
%matplotlib inline
from numpy.random import choice
import matplotlib.pyplot as plt
import numpy as np
```

```python
def CRP(N, alpha):
    
    if n < 1:
        return None
    
    assignments = np.empty(n) # number of table assignments
    next_table = 0
    
    for k in range(n):
        
        if np.random.random() < (1. * alpha / (alpha + k)):
            
            # visitor sits at new table
            assignments[k] = next_table
            next_table += 1
            
        else: 
            
            # selection prob as function of population
            probs = [(assignments[:k]==i).sum()/float(k) 
                     for i in range(next_table)]
            # randomly assign to existing table
            assignments[k] = choice(range(next_table), p=probs)
            
    return assignments
```

```python
n = 10
alpha = 1

def plot_crp(n_tables, ax=None):
    x = range(int(n_tables.max()) + 1)
    f = [(n_tables==i).sum() for i in set(n_tables)]
    if ax is None: ax = plt
    ax.bar(x, f)
    
fig, axes = plt.subplots(2, 5, sharex=True, sharey=True, figsize=(10,6))
for ax in np.ravel(axes):
    plot_crp(CRP(n, alpha), ax=ax)
```
